2. Crie um programa que verifique se uma letra digitada é "F" ou "M". Conforme a letra imprimir:
* F - Feminino;
* M - Masculino;
* Sexo Inválido.

```
sexo = str(input('Qual seu sexo, F ou M: '))
```

```
if(sexo == ('F' or 'f')):
  print('F - Feminino')
elif(sexo == ('M' or 'm')):
  print('M - Mascuino')
else:
  print('Sexo Inválido')
```
