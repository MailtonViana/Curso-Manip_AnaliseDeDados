#4. Correlação de Tabelas

**4.1 Agrupamento de dados**

df.groupby('coluna a ser agrupada']).agg({'coluna1':'sum','coluna2':'mean','coluna3':'max',...}).reset_index()

```
df_vendas = df_vendas.groupby(['material']).agg({'qtd_venda':'sum'}).reset_index()
```

```
df_vendas
```

```
df_pedidos = df_pedidos.groupby(['material']).agg({'qtd_estoque':'sum','qtd_pedido':'sum'}).reset_index()
```

```
df_pedidos
```

**4.2 Merge**

pd.merge(tabela_esquerda,tabela_direita, on = ['coluna em comum'], how = 'método')

metodo poe ser: left(esquerda), right(direita), inner(itens em comum), outer (todos os itens)

```
df_relatorio = pd.merge(df_materiais, df_pedidos, on = ['material'], how = 'left')
```

```
df_relatorio
```

```
df_relatorio = pd.merge(df_relatorio, df_vendas, on = ['material'], how = 'left')
```

**4.3 Tratamento pós merge**

```
df_relatorio[['qtd_estoque','qtd_pedido','qtd_venda']] = df_relatorio[['qtd_estoque','qtd_pedido','qtd_venda']].astype(int) 
```

#5. Regra de Negócio

Operações entre colunas sem condicional

```
df_relatorio['qtd_loja'] = df_relatorio['qtd_estoque'] + df_relatorio['qtd_pedido']
```

```
df_relatorio = df_relatorio[['material','qtd_loja','qtd_venda']].copy()
```

Operações entre colunas com condicional

```
def calcular_porcentagem(loja,venda):
    if (venda!=0):
        return loja/venda
    elif (venda==0) and (loja!=0):
        return 1.2
    else:
        return 0
```

df.apply(lambda row: funcao(row['coluna']), axis='columns')

```
df_relatorio['porcentagem'] = df_relatorio.apply(lambda row: calcular_porcentagem(row['qtd_loja'],row['qtd_venda']), axis = 'columns')
```

```
def verificar_status(porc):
  if (porc>=1.2):
    return 'alerta'
  else:
    return ''
```

```
df_relatorio['status'] = df_relatorio.apply(lambda row: verificar_status(row['porcentagem']),axis= 'columns')
```

#Filtro

df[(condicao)]

and -> & or -> |

```
df_relatorio = df_relatorio[(df_relatorio['status']!='alerta') & (df_relatorio['qtd_loja']!=0)]
```

#Ordenando Valores

df.sort_values(['colunas a serem ordenadas'], ascending=True) -> ordem crescente

df.sort_values(['colunas a serem ordenadas'], ascending=False) -> ordem decrescente

```
df_relatorio = df_relatorio.sort_values(['qtd_loja'], ascending=False)
```
