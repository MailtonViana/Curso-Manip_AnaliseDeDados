```
def par_ou_impar(numero):
    if(numero % 2 == 0):
      return 'Par'
    else:
      return 'Ímpar'
```

```
def calculadora_basica(num1,num2):
    soma = num1 + num2
    subtracao = num1 - num2
    multiplicacao = num1 * num2
    divisao = num1 / num2
    return soma, subtracao, multiplicacao, divisao
```

```
soma, sub, mult, div = calculadora_basica(4,2)
```

**Exercício 1. Faça um programa que receba como parâmetro o raio de um círculo e calcule a área.**

```
import math  ## Importa a bibioteca de Matemática
math.pi
```

```
def area_circulo(raio):
# area = pi * raio ^ 2
    area = math.pi * raio ** 2
    return round(area,4)
```

```
area_circulo(3)
```
